```python
def fillNaObjMode(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])

def fillNaIntMode(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])

def fillNaObjMode(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])

```

```python
def fill_missing_values_by_dtype(df):
    for col in df.columns:
        if df[col].isnull().any():  
            if df[col].dtype in ['float64', 'float32']:  
                df[col] = df[col].fillna(df[col].mean())
            elif df[col].dtype in ['int64', 'int32']:
                df[col] = df[col].fillna(df[col].mode()[0])
            elif df[col].dtype == 'object':  
                df[col] = df[col].fillna(df[col].mode()[0])

fill_missing_values_by_dtype(df)
```

```python
def fill_missing_values_by_dtype(test_data):
    for col in test_data.columns:
        if test_data[col].isnull().any():  
            if test_data[col].dtype in ['float64', 'float32']:  
                test_data[col] = test_data[col].fillna(test_data[col].mean())
            elif test_data[col].dtype in ['int64', 'int32']:
                test_data[col] = test_data[col].fillna(test_data[col].mode()[0])
            elif test_data[col].dtype == 'object':  
                test_data[col] = test_data[col].fillna(test_data[col].mode()[0])
```
